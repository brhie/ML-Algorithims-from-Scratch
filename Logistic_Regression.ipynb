```python
import numpy as np
import matplotlib.pyplot as plt
from math import e, log
np.set_printoptions(precision=100)
%matplotlib inline
```

```python
x = np.array([0, 1, 2, 4, 5, 6, 7, 11, 13])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1, 1])
plt.scatter(x, y)
plt.show()
```

```python
def logistic_function(z):
  return 1 / (1 + np.exp(-z))

z_example = np.arange(-4, 4, 0.01)

plt.plot(z_example, logistic_function(z_example))
plt.axvline(0, c="r")
plt.show()
```

```python
def loss_function(x_i, y_i, w, b):
  z = w * x_i + b
  pred = logistic_function(z)
  return (- y_i * np.log(pred)) - ((1 - y_i) * np.log(1 - pred))

def cost_function(x, y, w, b):
  m = len(x)
  cost = 0.0
  for i in range(m):
    cost += loss_function(x[i], y[i], w, b)
  return (1 / m) * cost

cost_function(x, y, 2, - 7)
```

```python
x_example = np.arange(0, 12, 0.01)
plt.scatter(x, y)
plt.plot(x_example, logistic_function(x_example * 2 - 7), c="r")
plt.show()
```

```python
def gradient_descent(x, y, w, b, n, alpha):
  m = len(x)
  for i in range(n):
    z = w * x + b
    gradient_w = (1 / m) * np.sum((logistic_function(z) - y) * y)
    gradient_b = (1 / m) * np.sum(logistic_function(z) - y)

    w = w - alpha * gradient_w
    b = b - alpha * gradient_b

    if i % 1000 == 0:
      print(f"iteration {i} -> w: {w}, b: {b}, cost: {cost_function(x, y, w, b)}")

    if i == n - 1:
      print(f"final value -> w: {w}, b: {b}, cost: {cost_function(x,y,w,b)}")
      return w, b
```

```python
w1, b1 = gradient_descent(x, y, 0, 0, 10000, 0.05)
```
